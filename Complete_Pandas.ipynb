**Creating a DataFrame**

A DataFrame is simply a table made up of multiple arrays. In the example below, the code would create a table with two columns, Code and Hub.

```
pd.DataFrame({'Code':[1,2,3], 'hub':[4,5,6]}, index=[1,2,3])
```

**Create a Series**

A Series is a sequence of values, also known as a list. From a visual perspective, imagine it being one column of a table

```
pd.Series(['Abhishek','Shubham','Riya'],index=[0,1,2], name ='Coders')
```
